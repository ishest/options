# Gamma scalping

Indeed, "gamma scalping" is an advanced trading strategy used mainly by option traders. It leverages the position's gamma to take advantage of price movements in the underlying asset, especially when the market is volatile. Here's a brief overview:

**Gamma Scalping:**

**Gamma and Delta:** First, let's understand two of the Greeks in options trading:
- **Delta:** Measures the sensitivity of the option's price to a $1 change in the price of the underlying asset. A delta of 0.50 means the option's price changes by 50 cents for every $1 change in the underlying.
- **Gamma:** Measures the rate of change in an option's delta for a $1 change in the price of the underlying asset. In simpler terms, gamma tells you how fast the delta is changing.

**How It Works:** A trader who has a straddle (long a call and long a put with the same strike price and expiration date) will benefit from large moves in either direction. This is because a straddle is a long volatility position with positive gamma. As the underlying's price moves, the delta of the position changes. If the trader constantly re-hedges to maintain a delta-neutral position, they can profit from these re-hedges if the underlying moves back and forth. This process of frequently adjusting the delta by buying or selling the underlying is called gamma scalping.

**Why "Scalping"?**: The term "scalping" traditionally refers to taking small profits frequently. In gamma scalping, the trader aims to capture small profits from the changes in delta and the subsequent adjustments (re-hedging) of their position.

**When is it Useful?:** Gamma scalping is especially useful when:
- The trader expects the underlying asset to be volatile but is unsure of the direction.
- The trader wants to offset the time decay (theta) of their options. By gamma scalping, they can earn profits that can potentially offset the loss from theta.

**Challenges:**
- Gamma scalping requires constant monitoring of the markets and frequent trading, which can result in higher transaction costs.
- It's a strategy that requires a deep understanding of option Greeks and their interplay.

It's worth noting that gamma scalping is an advanced strategy and is not suitable for everyone. It requires a solid understanding of options trading and the associated risks.




### Gamma Scalping Example

### Initial Position

Imagine you believe that stock **XYZ**, currently trading at **$100**, will experience significant volatility over the next month, but you're unsure in which direction it will move.

You decide to buy a straddle:
- Long one $100 strike call option for a premium of **$5**.
- Long one $100 strike put option for a premium of **$5**.
- Your total premium outlay is **$10** ($5 + $5).

### Understanding the Position

Since you have a straddle at-the-money:
- The delta of the call option might be around **0.5**.
- The delta of the put option might be around **-0.5**.
  
Thus, your net delta is roughly zero (0.5 - 0.5 = 0). This means your position is **delta neutral** at initiation. However, your position has **positive gamma**. If stock XYZ moves up or down, the delta will change, potentially making your position no longer delta neutral.

### Stock Movement

The next day, stock **XYZ** jumps to **$102**. Due to the positive gamma:
- Your call option's delta might increase to **0.6**.
- Your put option's delta might decrease to **-0.4**.

This means your net delta is now **0.2**, and you're no longer delta neutral.

## Re-hedging

To return to a delta-neutral position, you'd sell **20 shares** of stock XYZ (since net delta is 0.2, equivalent to 20 shares for 1 option contract). By doing this, you lock in a small profit from the stock's upward move.

## Subsequent Stock Movement

Now, let's say **XYZ** drops back to **$100** the following day. Because of your positive gamma:
- The call option's delta drops back to around **0.5**.
- The put option's delta returns to about **-0.5**.

Your net delta changes again due to the short stock position.

### Re-hedging Again

You'd buy back the **20 shares** of XYZ you previously sold to return to a delta-neutral position, capturing another small profit, this time from the stock's downward move.

This explanation is simplified, but the process involves constantly adjusting your stock position to remain delta neutral. Gamma scalping enables you to lock in small profits from these re-hedges, ideally compensating for the time decay (theta) and potentially the initial premium outlay.

### Considerations in Real Trading

- **Transaction Costs**: Frequent adjustments can lead to significant transaction costs.
- **Monitoring**: This strategy requires constant monitoring. Automated systems or trading algorithms are often preferred.
- **Risks**: Low volatility can lead to losses due to time decay. Gamma scalping thrives in volatile environments.






```python
import matplotlib.pyplot as plt

stock_price = 100
call_premium = 5
put_premium = 5
total_premium = call_premium + put_premium
total_premium
```

```python
# Python delta setup
call_delta = 0.5
put_delta = -0.5
net_delta = call_delta + put_delta
net_delta
```

```python
# Stock price change
new_stock_price = 102

# Delta adjustments after stock price change
new_call_delta = 0.6
new_put_delta = -0.4
new_net_delta = new_call_delta + new_put_delta

# Re-hedging
shares_to_sell = 20  # based on net delta of 0.2 for 1 option contract

# Plotting
```

```python
# 1. Stock Price Movements
plt.figure(figsize=(10, 6))
plt.plot(['Initial', 'After Change'], [stock_price, new_stock_price], marker='o', color='blue')
plt.title('Stock Price Movements')
plt.ylabel('Stock Price')
plt.grid(True)
plt.show()

# 2. Delta Changes Visualization
plt.figure(figsize=(10, 6))
plt.plot(['Initial', 'After Change'], [call_delta, new_call_delta], label='Call Delta', marker='o', color='green')
plt.plot(['Initial', 'After Change'], [put_delta, new_put_delta], label='Put Delta', marker='o', color='red')
plt.plot(['Initial', 'After Change'], [net_delta, new_net_delta], label='Net Delta', marker='o', linestyle='--', color='blue')
plt.title('Delta Changes with Stock Price Movement')
plt.ylabel('Delta Value')
plt.legend()
plt.grid(True)
plt.show()

# 3. Re-hedging Process
plt.figure(figsize=(10, 6))
plt.bar(['Shares to Sell'], [shares_to_sell], color='purple')
plt.title('Re-hedging Process (Shares Sold)')
plt.ylabel('Number of Shares')
plt.grid(axis='y')
plt.show()
```
